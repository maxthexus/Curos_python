```python
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
```

```python
np.loadtxt(url,delimiter=',',usecols=np.arange(1,6,1),skiprows=1)
```

```python
dados = np.loadtxt(url,delimiter=',',usecols=np.arange(1,6,1),skiprows=1)
```

```python
diametro_laranja = dados[:5000,0]
diametro_toranja = dados[5000:,0]
peso_laranja = dados[:5000,1]
peso_toranja = dados[5000:,1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(peso_laranja,diametro_laranja)
```

```python
plt.plot(peso_toranja,diametro_toranja)
```

```python
Y = peso_laranja
X = diametro_laranja
n = np.size(X)

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X+b
```

```python
plt.plot(peso_laranja,diametro_laranja)
plt.plot(y,X)
```
